# Линейная регрессия

Теоретическая справка

- Математическое ожидание
  
$$E[X] = \frac{1}{n}\sum_{i=1}^{n} x_i$$

- Дисперсия
  
$$Var[X] = (E[X^2]) - (E[X])^2$$

- Ковариация
  
$$cov(X, Y) = E[XY] - E[X] \cdot E[Y]$$


Теоретическая справка

Модель линейной регрессии
Для одномерного случая (одна фича):

$$y = w_1 \cdot x + w_0$$
где:
- $w_1$ - угловой коэффициент
- $w_0$ - свободный член

Формулы для вычисления весов

**Угловой коэффициент:**
$$w_1 = \frac{cov(X, Y)}{var(X)} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$

**Свободный член:**

$$w_0 = \bar{y} - w_1 \cdot \bar{x}$$

# SVD сжатие изображений

Сингулярное разложение (SVD)
Любую матрицу A можно представить в виде:
$$A = U × S × V^T$$

Где:
- $U$ - левые сингулярные векторы
- $S$ - диагональная матрица сингулярных чисел
- $V^T$ - правые сингулярные векторы (транспонированные)

Самое интересное, это матрица $S$, ведь она содержит основную информацию о картинке, и если обрезать часть этой матрицы, то мы сможем сжимать отображения или по другому:

Сохраняя только k наибольших сингулярных чисел, мы получаем приближенную матрицу:
$$A_k = U[:, :k] × S[:k, :k] × V^T[:k, :]$$

## Декоратор

```python
def benchmark_comparison(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        
        execution_time = end_time - start_time
        print(f"Функция {func.__name__} выполнилась за {execution_time:.6f} секунд")
        
        return result, execution_time
    return wrapper

@benchmark_comparison
def numpy_mean(arr):
    return np.mean(arr)

@benchmark_comparison
def python_mean(arr):
    return sum(arr) / len(arr)

@benchmark_comparison
def numpy_create_matrix(n):
    return np.random.rand(n, n)

@benchmark_comparison
def python_create_matrix(n):
    return [[random.random() for _ in range(n)] for _ in range(n)]

@benchmark_comparison
def numpy_has_zero_columns(matrix):
    return np.any(np.all(matrix == 0, axis=0))

@benchmark_comparison
def python_has_zero_columns(matrix):
    if not matrix or not matrix[0]:
        return False
    
    n_cols = len(matrix[0])
    n_rows = len(matrix)
    
    for j in range(n_cols):
        all_zero = True
        for i in range(n_rows):
            if matrix[i][j] != 0:
                all_zero = False
    if all_zero:
        return True
    return False

@benchmark_comparison
def numpy_diagonal_sum(matrix):
    return np.trace(matrix)

@benchmark_comparison
def python_diagonal_sum(matrix):
    total = 0
    n = min(len(matrix), len(matrix[0])) if matrix else 0
    for i in range(n):
        total += matrix[i][i]
    return total

def compare_performance():
    arr = np.arange(100000)
    numpy_result, numpy_time = numpy_mean(arr)
    python_result, python_time = python_mean(arr.tolist())
    print(f'Ускорение: {python_time / numpy_time}')

    numpy_result, numpy_time = numpy_create_matrix(10000)
    python_result, python_time = python_create_matrix(10000)
    print(f'Ускорение: {python_time / numpy_time}')

    numpy_matrix = np.random.rand(10000, 10000)
    python_matrix = numpy_matrix.tolist()
    
    zero_col_idx = 10000 // 2
    numpy_matrix[:, zero_col_idx] = 0
    for i in range(10000):
        python_matrix[i][zero_col_idx] = 0

    numpy_result, numpy_time = numpy_has_zero_columns(numpy_matrix)
    python_result, python_time = python_has_zero_columns(python_matrix)
    print(f'Ускорение: {python_time / numpy_time}')
    
    python_result, python_time = numpy_diagonal_sum(numpy_matrix)
    python_result, python_time = python_diagonal_sum(python_matrix)
    print(f'Ускорение: {python_time / numpy_time}')
```

```python
compare_performance()
```

## Лин рег

```python
def mean(values):
    return np.sum(values * 1/len(values))
```

```python
def variance(values):
    ex = mean(values)
    ex2 = mean(values * values)
    return ex2 - ex ** 2
```

```python
def covariance(x, y):
    ex = mean(x)
    ey = mean(y)
    exy = mean(x * y)
    return exy - ex * ey
```

```python
def compute_linear_regression_weights(X, y):
    """
    Вычисляет веса для одномерной линейной регрессии
    
    Параметры:
    X: numpy array
        Входные признаки (одномерный массив)
    y: numpy array
        Целевые значения
        
    Возвращает:
    tuple: (w0, w1) - свободный член и угловой коэффициент
    """
    cov_xy = covariance(X, y)
    var_x = variance(X)
    
    w1 = cov_xy / var_x
    w0 = mean(y) - w1 * mean(X)
    
    return w0, w1
```

```python
def predict(X, w0, w1):
    """
    Выполняет предсказание по модели линейной регрессии
    
    Параметры:
    X: numpy array
        Входные признаки
    w0: float
        Свободный член
    w1: float
        Угловой коэффициент
        
    Возвращает:
    numpy array: предсказанные значения
    """
    return w1 * X + w0
```

```python
def generate_dataset():
    np.random.seed(42)
    n_points = 100
    X = np.linspace(0, 10, n_points)
    y_true = 2 * X + 1
    y = y_true + np.random.normal(0, 1, n_points)
    return X, y


X, y = generate_dataset()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(X, y, alpha=0.7, label='Данные')

X_line = np.linspace(X.min(), X.max(), 100)
w0, w1 = compute_linear_regression_weights(X, y)
y_line = predict(X_line, w0, w1)
plt.plot(X_line, y_line, 'r-', linewidth=2, label=f'Регрессия: y = {w1:.3f}x + {w0:.3f}')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Линейная регрессия')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Жизнь

```python
class GameOfLife:
    def __init__(self, size=50):
        self.size = size
        self.grid = np.random.choice([0, 1], size=(size, size), p=[0.8, 0.2])
        
    def count_neighbors(self, grid):
        """Подсчет количества живых соседей для каждой клетки"""
        kernel = np.array([[1, 1, 1],
                          [1, 0, 1],
                          [1, 1, 1]])
        return convolve2d(grid, kernel, mode='same', boundary='wrap')
    
    def update(self):
        """Обновление состояния игры по правилам"""
        neighbors = self.count_neighbors(self.grid)
        
        # Правила игры:
        # 1. Живая клетка с 2 или 3 соседями выживает
        # 2. Мертвая клетка с 3 соседями оживает
        # 3. Остальные умирают или остаются мертвыми
        
        birth = (self.grid == 0) & (neighbors == 3)
        survive = (self.grid == 1) & ((neighbors == 2) | (neighbors == 3))
        
        self.grid = np.where(birth | survive, 1, 0)
        return self.grid


# Функция для свертки (если scipy нет)
def convolve2d(grid, kernel, mode='same', boundary='wrap'):
    """Реализация 2D свертки"""
    size = grid.shape[0]
    result = np.zeros_like(grid)
    
    for i in range(size):
        for j in range(size):
            total = 0
            for ki in range(-1, 2):
                for kj in range(-1, 2):
                    if kernel[ki+1, kj+1] == 0:
                        continue
                    ni = (i + ki) % size
                    nj = (j + kj) % size
                    total += grid[ni, nj]
            result[i, j] = total
    return result

def run_simulation(steps=100, size=50):
    """Запуск симуляции на определенное количество шагов"""
    game = GameOfLife(size=size)
    
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.flatten()
    
    for step in range(steps):
        if step < 10:
            axes[step].imshow(game.grid, cmap='binary')
            axes[step].set_title(f'Шаг {step}')
            axes[step].axis('off')
        game.update()
    
    plt.tight_layout()
    plt.show()


run_simulation()
```
